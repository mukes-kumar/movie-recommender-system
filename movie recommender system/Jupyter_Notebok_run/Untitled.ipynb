```python
import ast

def covert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L


# ast ye ek modul hota hai jo int type ke dic or sare dic. ko list me karta hai
```

```python
movies['genres'] = movies['genres'].apply(covert)

# genres column me se name nikal jayega..............
```

```python
movies
```

```python
movies['keywords'] = movies['keywords'].apply(covert)
```

```python
movies
```

```python
# cast column me se kuch name information nikalne ke liye ..................
def convert3(obj):
    L=[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
# crew column me se kuch name information nikalne ke liye .................

def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies
```

```python
# overview column me string hai isliye ese list me split karen ge.........
# we can see that..
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies
```

```python
# # genres , cast , crew and other column me kisi bhi name bich me space hai to
# # usko remove karna taki koi searching me koi dikat na aaye ..

# # 'Sam Worthington'
# # 'SamWorthington'

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# To create new columns 
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
# !pip install nltk
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```
